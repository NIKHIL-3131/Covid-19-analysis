```python
Confirmed1 = pd.melt(Confirmed, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                           var_name=['Date'])

Recovered1 = pd.melt(Recovered, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                           var_name=['Date'])

Deaths1 = pd.melt(Deaths, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                           var_name=['Date'])
```

```python
Confirmed1.shape
```

```python
Confirmed1.head(5)
```

```python
Confirmed1['Date'] = pd.to_datetime(Confirmed1['Date'])
Recovered1['Date'] = pd.to_datetime(Recovered1['Date'])
Deaths1['Date'] = pd.to_datetime(Deaths1['Date'])
```

```python
Confirmed1.columns = Confirmed1.columns.str.replace('value', 'Confirmed')
Recovered1.columns = Recovered1.columns.str.replace('value', 'Recovered')
Deaths1.columns = Deaths1.columns.str.replace('value', 'Deaths')

```

```python
Confirmed1.head(5)
```

```python
Confirmed1.isnull().sum()
```

```python
Deaths1.isnull().sum()
```

```python
Confirmed1['Province/State'].fillna(Confirmed1['Country/Region'], inplace=True)
Recovered1['Province/State'].fillna(Recovered1['Country/Region'], inplace=True)
Deaths1['Province/State'].fillna(Deaths1['Country/Region'], inplace=True)
```

```python
Recovered1.isnull().sum()
```

```python
Deaths1.isnull().sum()
```

```python
Deaths1['Lat'].fillna(Deaths1['Country/Region'], inplace=True)
Deaths1['Long'].fillna(Deaths1['Country/Region'], inplace=True)
Recovered1['Lat'].fillna(Recovered1['Country/Region'], inplace=True)
Recovered1['Long'].fillna(Recovered1['Country/Region'], inplace=True)
Confirmed1['Lat'].fillna(Confirmed1['Country/Region'], inplace=True)
Confirmed1['Long'].fillna(Confirmed1['Country/Region'], inplace=True)
```

```python
Join = Confirmed1.merge(Deaths1[['Province/State','Country/Region','Date','Deaths']], 
                                      how = 'left', 
                                      left_on = ['Province/State','Country/Region','Date'], 
                                      right_on = ['Province/State', 'Country/Region','Date'])
Join.head()
```

```python
Join = Join.merge(Recovered1[['Province/State','Country/Region','Date','Recovered']], 
                                      how = 'left', 
                                      left_on = ['Province/State','Country/Region','Date'], 
                                      right_on = ['Province/State', 'Country/Region','Date'])
Join.head()
```

```python
Join['Month-Year'] = Join['Date'].dt.strftime('%b-%Y')
```

```python
Join.head(5)
```

```python
Join1 = Join.copy()

#creating a new date columns - 1
Join1['Date - 1'] = Join1['Date'] + pd.Timedelta(days=1)
Join1.rename(columns={'Confirmed': 'Confirmed - 1', 'Deaths': 'Deaths - 1', 'Recovered': 'Recovered - 1',
                          'Date': 'Date Minus 1'}, inplace=True)
```

```python
Join1.head(5)
```

```python
Join2 = Join.merge(Join1[['Province/State', 'Country/Region','Confirmed - 1', 'Deaths - 1', 
                            'Recovered - 1', 'Date - 1', 'Date Minus 1']], how = 'left',
                             left_on = ['Province/State','Country/Region','Date'], 
                             right_on = ['Province/State', 'Country/Region','Date - 1'])
Join2.head()
```

```python
Join2['Confirmed Daily'] = Join2['Confirmed'] - Join2['Confirmed - 1']
Join2['Deaths Daily'] = Join2['Deaths'] - Join2['Deaths - 1']
Join2['Recovered Daily'] = Join2['Recovered'] - Join2['Recovered - 1']

print(Join2.shape)
```

```python

#Braking the numbers by Day


#creating a new dataframe   
Join1 = Join.copy()

#creating a new date columns - 1
Join1['Date - 1'] = Join1['Date'] + pd.Timedelta(days=1)
Join1.rename(columns={'Confirmed': 'Confirmed - 1', 'Deaths': 'Deaths - 1', 'Recovered': 'Recovered - 1',
                          'Date': 'Date Minus 1'}, inplace=True)

#Joing on the 2 DFs
Join2 = Join.merge(Join1[['Province/State', 'Country/Region','Confirmed - 1', 'Deaths - 1', 
                            'Recovered - 1', 'Date - 1', 'Date Minus 1']], how = 'left',
                             left_on = ['Province/State','Country/Region','Date'], 
                             right_on = ['Province/State', 'Country/Region','Date - 1'])

#minus_onedf.rename(columns={'Confirmed': 'Confirmed - 1', 'Deaths': 'Deaths - 1', 'Recovered': 'Recovered - 1'}, inplace=True)

Join2.head()

# Additional Calculations

Join2['Confirmed Daily'] = Join2['Confirmed'] - Join2['Confirmed - 1']
Join2['Deaths Daily'] = Join2['Deaths'] - Join2['Deaths - 1']
Join2['Recovered Daily'] = Join2['Recovered'] - Join2['Recovered - 1']

print(Join2.shape)
```

```python
# Adding manually the numbers for first day

Join2['Confirmed Daily'].loc[Join2['Date'] == '2020-01-22'] = Join2['Confirmed']
Join2['Deaths Daily'].loc[Join2['Date'] == '2020-01-22'] = Join2['Deaths']
Join2['Recovered Daily'].loc[Join2['Date'] == '2020-01-22'] = Join2['Recovered']

# deleting columns
del Join2['Confirmed - 1']
del Join2['Deaths - 1']
del Join2['Recovered - 1']
del Join2['Date - 1']
del Join2['Date Minus 1']
```
